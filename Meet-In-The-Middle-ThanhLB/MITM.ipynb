```python
from random import random

N_ = 20
W_ = [int(random() * 100) + 1 for x in range(20)]
P_ = [int(random() * 100) + 1 for x in range(20)]
C_ = 500000
```

```python
# Python program to demonstrate working of Meet in the
# Middle algorithm for maximum subset sum problem.
from random import random
from time import time
from typing import List
import bisect
X = []
X_p = [] 
Y = []
Y_p = []

# Find all possible sum of elements of a[] and store
# in x[]
def calcsubarray(a: List[int], b : List[int], x: List[int], x_p : List[int], n: int, c: int) -> None:
    for i in range(2 ** n):
        s = 0
        p = 0
        for j in range(n):
            if i & 2 ** j:
                s += a[j + c]
                p += b[j + c]
        x.append(s)
        x_p.append(p)

def sort(X : List[int], X_p : List[int]) -> None:
    for i in range(len(X) - 1):
        for j in range(i + 1, len(X)):
            if X[i] > X[j]:
                temp = X[i]
                X[i] = X[j]
                X[j] = temp

                temp = X_p[i]
                X_p[i] = X_p[j]
                X_p[j] = temp

# Returns the maximum possible sum less or equal to S
def MITM(W, wt: List[int], val : List[int], n: int) -> int:
	global Y
	
	# Compute all subset sums of first and second
	# halves
	calcsubarray(wt, val, X, X_p, n // 2, 0)
	calcsubarray(wt, val, Y, Y_p, n - n // 2, n // 2)
	size_X = 2 ** (n // 2)
	size_Y = 2 ** (n - n // 2)

	# Sort Y (we need to do doing binary search in it)
	sort(Y, Y_p)

	# To keep track of the maximum sum of a subset
	# such that the maximum sum is less than S
	maxx = 0
	maxp = 0
	# Traverse all elements of X and do Binary Search
	# for a pair in Y with maximum sum less than S.
	for i in range(size_X):

		if (X[i] <= W):

			# lower_bound() returns the first address
			# which has value greater than or equal to
			# S-X[i].
			p = bisect.bisect_left(Y, W - X[i])

			# If S-X[i] was not in array Y then decrease
			# p by 1
			if (p == size_Y or (p < size_Y and Y[p] != (W - X[i]))):
				p -= 1
                
			if ((Y[p] + X[i]) > maxx and Y_p[p] + X_p[i] > maxp):
				maxx = Y[p] + X[i]
				maxp = Y_p[p] + X_p[i]
	return maxp
```

```python
# code
# A Dynamic Programming based Python
# Program for 0-1 Knapsack problem
# Returns the maximum value that can
# be put in a knapsack of capacity W
 
 
from random import random


def DP(W, wt, val, n):
    dp = [0 for i in range(W+1)]  # Making the dp array
 
    for i in range(1, n+1):  # taking first i elements
        for w in range(W, 0, -1):  # starting from back,so that we also have data of
                                # previous computation when taking i-1 items
            if wt[i-1] <= w:
                # finding the maximum value
                dp[w] = max(dp[w], dp[w-wt[i-1]]+val[i-1])
 
    return dp[W]  # returning the maximum value of knapsack
 
 
```

```python
from time import time
# Driver code
val = P_
wt = W_
W = C_
n = N_
# This code is contributed by Suyash Saxena
print("\nDynamic Programing\n--------------------")
t1 = time()
print("Optimized result: {}".format(DP(W, wt, val, n)))
t2 = time()
print("Time with C = {}, N = {}: {}".format(W, N_, t2 - t1))

print("\nMeet-In-The-Middle\n--------------------")
t1 = time()
print("Optimized result: {}".format(MITM(W, wt, val, n)))
t2 = time()
print("Time with C = {}, N = {}: {}".format(W, N_, t2 - t1))
```
